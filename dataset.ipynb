```python
import os
```

```python

folder_path = r'Dataset\NotEngagement\drowsy'

# Iterate through the files in the folder
for filename in os.listdir(folder_path):
    # Check if the file is an image (you can use more specific checks if needed)
    if filename.endswith(('.jpg', '.jpeg', '.png', '.gif')):
        # Construct the new name by adding "_12" to the end
        new_name = os.path.join(folder_path, filename[:-4] + '_1' + filename[-4:])
        old_name = os.path.join(folder_path, filename)
        
        # Rename the file
        os.rename(old_name, new_name)
        print(f'Renamed: {filename} -> {new_name}')
```
